```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
```

```python
dataset
```

```python
dataset["classification_yes"].value_counts()
```

```python
dataset.columns 
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes',]]
```

```python
independent.shape
```

```python
independent
```

```python
dependent=dataset[['classification_yes']]
```

```python
dependent.shape
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
x_train
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB
param_grid={"var_smoothing": [1e-10, 1e-9, 1e-8],
    "priors": [None, [0.7, 0.3]]}
grid=GridSearchCV(GaussianNB(), param_grid, refit=True, verbose=3, n_jobs=-1, scoring="accuracy")
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
```

```python
print(re)
```

```python
format(grid.best_params_)
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
grid_prediction=grid.predict(x_test)
```

```python
grid_prediction
```

```python
from sklearn.metrics import confusion_matrix
cfm=confusion_matrix(grid_prediction,y_test)
```

```python
print(cfm)
```

```python
from sklearn.metrics import classification_report 
clf_report=classification_report(y_test,grid_prediction)
```

```python
print(clf_report)
```
