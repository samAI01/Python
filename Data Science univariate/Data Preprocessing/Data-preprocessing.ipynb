```python
import pandas as pd
```

```python
dataset=pd.read_csv("Placement.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.fillna({"salary":0},inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset["salary"].fillna(dataset["salary"].mean())
```

```python
dataset
```

```python
def quaQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        #print(columnName)
        if (dataset[columnName].dtype=='str'):
           # print("qual")
            qual.append(columnName)
        else:
           # print("quan")
            quan.append(columnName)
    return quan,qual
```

```python
quaQual(dataset)
```

```python
quan,qual=quaQual(dataset)
```

```python
dataset[quan]
```

```python
import numpy as np
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy='constant')
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
df=pd.DataFrame(df,columns=quan)
```
