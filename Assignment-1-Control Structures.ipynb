```python
#Enter the BMI Index:34
Underweight: Below 18.5
Normal weight: 18.5 – 24.9
Very Overweight: 25.0 – 29.9
Obese: 30.0 and above 
```

```python
lists=[18.5,24.9,29.9,30]
for (weight) in lists:
    if(weight<=18.5):
        print("Under weight")
    elif(weight<=24.9):
        print("Normal weight")
    elif(weight<=29.9):
        print("Very Overweight")
    else:
        print("Obese")

    
```
